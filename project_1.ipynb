```python
import matplotlib.pyplot as plt
import numpy as np

Vs = (np.loadtxt("Linder_S_finamod.txt"))
Vp = (np.loadtxt("Linder_P_finamod.txt"))
grid = (np.loadtxt("grid_cell_centers.txt"))
x = grid[:,0]
y = grid[:,1]
nrows = 3
ncols = 10
# Calculation of Allowable Bearing Presure (Qa) from Vp and Vs Dataset
gam_0 = 16
n = 4
gam = gam_0 + 0.002*Vp
Qa = 0.1*gam*Vs/n
np.savetxt("Linder_Qa_finamod.txt", Qa, fmt='%.18e', delimiter=' ', newline='\n', header='', footer='')
# Visualization of P-Wave Velocity Dataset
fig1, axes = plt.subplots(nrows,ncols,figsize=(20,10))
for ax,n in zip(axes.flat,range(0,30)):
    dp = ax.tricontourf(x,y,Vp[:,n],cmap='jet')
    ax.set_title('P%r'%(n+1))
    ax.set(xlabel='x [m]', ylabel='y [m]')
    ax.label_outer()
cbar = fig1.colorbar(dp, ax=axes, location='bottom',shrink=0.3,pad=0.08)
cbar.set_label('P-Wave Velocity (m/s)')
# Visualization of S-Wave Velocity Dataset
fig2, axes = plt.subplots(nrows,ncols,figsize=(20,10))
for ax,n in zip(axes.flat,range(0,30)):
    dp = ax.tricontourf(x,y,Vs[:,n],cmap='jet')
    ax.set_title('S%r'%(n+1))
    ax.set(xlabel='x [m]', ylabel='y [m]')
    ax.label_outer()
cbar = fig2.colorbar(dp, ax=axes, location='bottom',shrink=0.3,pad=0.08)
cbar.set_label('S-Wave Velocity (m/s)')
# Visualization of Allowable Bearing Pressure Dataset 
fig, axes = plt.subplots(nrows,ncols,figsize=(20,10),)
for ax,n in zip(axes.flat,range(0,30)):
    dp = ax.tricontourf(x,y,Qa[:,n],cmap='jet')
    ax.set_title('Qa%r'%(n+1))
    ax.set(xlabel='x [m]', ylabel='y [m]')
    ax.label_outer()
cbar = fig3.colorbar(dp, ax=axes, location='bottom',shrink=0.3,pad=0.08)
cbar.set_label('Allowable Bearing Pressure')
#plt.show()
```
